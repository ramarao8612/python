Right Triangle Star Pattern

```python
n = 5

for i in range(1, n + 1):
    print("*" * i)
```

PYRAMID STAR PATTERN

```python
n = 5

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

Inverted Right Triangle

```python
n = 5

for i in range(n, 0, -1):
    print("*" * i)
```

NUMBER INCREASING PATTERN

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

DIAMOND PATTERN

```python
n = 5

# upper part
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))

# lower part
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
