Type casting: 
1.converting int to float to str

```python
# Type Casting Examples
x = "10"
y = int(x)   # string to int
z = float(x) # string to float

a = 5
b = str(a)   # int to string

print(y, type(y))
print(z, type(z))
print(b, type(b))
```

```python
# Example:User input always comes as string
age = "21"

# Convert to int to do math
age = int(age)
print(age + 5)   # 26

# Float example (marks/percentage)
marks = "89.5"
marks = float(marks)
print(marks + 0.5)   # 90.0

# Convert number to string (for printing)
num = 100
message = "Your score is " + str(num)
print(message)
```

for loop: we use for loop when looping through a particular given range. 
examples

```python
# Example 1: Print even numbers
for i in range(1, 11):
    if i % 2 == 0:
        print(i)

# Example 2: Sum of numbers
total = 0
for i in range(1, 6):
    total += i
print("Sum:", total)   

# Example 3: Loop through list
prices = [100, 200, 300]

for price in prices:
    print("Price with tax:", price * 2) 
    
#printing multiplication table. 
num = 5

for i in range(1, 11):
    print(num, "x", i, "=", num * i) 

#print sum of even numbers.
total = 0
for i in range(1, 21):
    if i % 2 == 0:
        total += i
print("Sum of even numbers:", total)

#count characters in the string. 
text = "python"
count = 0
for ch in text:
    count += 1
print("Length:", count) 

#find largest number in a list 
nums = [12, 45, 7, 89, 23]
max_val = nums[0]
for n in nums:
    if n > max_val:
        max_val = n
print("Max:", max_val) 

#filter data
marks = [45, 67, 80, 30, 90]
for m in marks:
    if m >= 50:
        print("Pass:", m) 
        
#pattern printing. 
for i in range(1, 6):
    print("*" * i)



```

while loop:A while loop is a programming control structure that repeatedly executes a block of code as long as a specified Boolean condition remains true.

```python
#user login system 
correct_password = "admin123"
entered = ""

while entered != correct_password:
    entered = input("Enter password: ")

print("Login Successful")
```

ATM Withdrawal using while loop.


```python
balance = 1000

while True:
    amount = int(input("Enter amount to withdraw: "))

    if amount <= balance:
        print("Withdraw successful")
        break
    else:
        print("Insufficient balance, try again")
```

```python
#number game.
secret = 7
guess = 0

while guess != secret:
    guess = int(input("Guess number: "))

    if guess < secret:
        print("Too small")
    elif guess > secret:
        print("Too big")

print("Correct!")
```

```python
#sum until user stops
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))

    if num == 0:
        break

    total += num

print("Total:", total)
```

```python
#reverse of number
num = 1234
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed:", rev)
```

```python
#counts digits in a number.
num = 98765
count = 0

while num > 0:
    count += 1
    num //= 10

print("Digits:", count)
```
